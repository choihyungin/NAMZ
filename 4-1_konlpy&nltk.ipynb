## 1. KoNLPy / NLTK

```python
from konlpy.tag import Twitter ; twitter = Twitter()
import nltk
```

## 2. Slove font error

```python
from matplotlib import font_manager, rc
font_fname = '/Library/Fonts/AppleGothic.ttf'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family = font_name)
```

## 3. Open txt files

```python
pos_sents = open("positive_sents.txt").read()[:4500000]
neg_sents = open("negative_sents.txt").read()
```

## 4. Positive sentences

```python
pos_sent = twitter.pos(pos_sents, norm = True, stem = True) #pos tagging
rm_formal_morphs = [i for i in pos_sent if i[1] != "Josa" and i[1] != "Punctuation"] #remove formal morphs and punctuation

stopwords_list = ["", "", "", ""]
rm_stopwords = [w for w in rm_formal_morphs if w[0] not in stopwords_list] #remove stopwords

fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

import matplotlib.pyplot as plt
plt.figure(figsize=(30, 10))
fdist.plot(100)      #top 100 words plotting
```

## 5. Negative sentences

```python
neg_sent = twitter.pos(neg_sents, norm = True, stem = True) #pos tagging
rm_formal_morphs = [i for i in neg_sent if i[1] != "Josa" and i[1] != "Punctuation"] #remove formal morphs and punctuation

stopwords_list = ["", "", "", ""]
rm_stopwords = [w for w in rm_formal_morphs if w[0] not in stopwords_list] #remove stopwords

fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

import matplotlib.pyplot as plt
plt.figure(figsize=(30, 10))
fdist.plot(100)      #top 100 words plotting
```
